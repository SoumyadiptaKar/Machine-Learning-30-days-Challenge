# Black Friday Dataset Exploratory Data Analysis

### Problem Statement
A retail company "ABC Private Limited" wants to understand the customer purchase behaviour (specifically, purchase amount) against various products of different categories. They have shared purchase summary of various customers for selected high volume products from last month. The data set also contains customer demographics (age, gender, marital status, city type, stay_in_current_city), product details (product_id and product category) and Total purchase_amount from last month.
Now, they want to build a model to predict the purchase amount of customer against various products which will help them to create personalized offer for customers against different products.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df_train=pd.read_csv('Black_Friday_Dataset/train.csv')
df_train.head()
```

```python
df_test=pd.read_csv('Black_Friday_Dataset/test.csv')
df_test.head()
```

```python
# Firstly merge both the test and train data.
df=df_train.append(df_test)
```

```python
# Lets convert the categorical features into numerical
pd.get_dummies(df['Gender']) # One way
```

```python
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
df['Age'].unique()
```

```python
pd.get_dummies(df['Age']) # This is one way but lets use some ordinal encoding by providing rank
# This way my map model will be able to understand the pattern (Target Guiding)
# Pattern in the sence that people within the range 18 to 50 will have more num of orders than the rest. 
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
# Fixing categorical City_Category
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
```

```python
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
# drop City_Category
df.drop('City_Category',axis=1,inplace=True)
```
