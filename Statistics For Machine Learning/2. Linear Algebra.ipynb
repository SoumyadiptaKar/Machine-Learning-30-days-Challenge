# Linear Algebra

Linear algebra is the domain of mathematics concerning linear equations and their representation in vector spaces and through matrices

```python
import numpy as np
```

```python
a=np.array([[10,5]])
b=np.array([[5,7]])
print(a)
print(b)
```

```python
# Addition of vectors
a+b
```

```python
a.T+b.T
```

```python
(a+b).T
```

```python
# Subtraction of vectors
a-b
```

```python
(a-b).T
```

```python
# Scalar Multiplication
a=2
u=np.array([[30,15]])

(a*u)
```

```python
(a*u).T
```

```python
# Dot product of 2 vectors
# For 2-D vectors,

u=np.array([10, 20])
v=np.array([25, 6])

print(10*25 + 20*6)
np.dot(u,v)
```

```python
a=[[5,10],[15,20]]
b=[10,12]

np.dot(a,b)
```

```python
a=[[5,10],[15,20],[25,30]]
b=[[10,12],[5,10]]

np.dot(a,b)
```

```python
# Identity Metrix

I1=np.eye(1)
I2=np.eye(2)
I3=np.eye(3)

print(I1,end=" = I1\n\n")
print(I2,end=" = I2\n\n")
print(I3,end=" = I3\n\n")
```

## Matrix Multiplication

```python
a=np.array([
    [1,4],
    [6,3]
])

b=np.array([
    [2,3],
    [3,5]
])

np.matmul(a,b)
```

```python
# We Know That AA^-1 = I 
a=np.array([
    [2,1,0],
    [0,1,0],
    [1,2,1]
])
a
```

```python
a_inv=np.linalg.inv(a)
a_inv
```

```python
np.matmul(a,a_inv)
```
