# Seaborn Tutorial
Seaborn is a Python data visualization library based on Matplotlib. 

It provides a high-level interface for creating attractive and informative statistical graphics. 

Seaborn is built on top of Matplotlib and integrates well with Pandas data structures, making it particularly useful for exploring and visualizing datasets.

## Distribution Plots :
1. distplot
2. joinplot
3. pairplot

```python
from sklearn import datasets

# using iris dataset for this tutorial
iris = datasets.load_iris()
```

```python
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame(
    iris.data, 
    columns=iris.feature_names
    )
```

```python
df.head()
```

### Corelation with Heatmap
A correlation heatmap uses colored cells, typically in a monochromatic scale, to show a 2D correlation matrix (table) between two discrete dimensions or event types. 

It is very important in Feature Selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

### JoinPlot
A join plot allows to study the relationship between 2 numeric variables. 

The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

```python
sns.jointplot(x='sepal length (cm)',y='petal length (cm)',data=df,kind='hex')
```

```python
sns.jointplot(x='sepal length (cm)',y='petal length (cm)',data=df,kind='reg')
```

### Pair Plot
A “pairs plot” is also known as a scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just elaborations on this, showing all variables paired with all the other variables

```python
sns.pairplot(df,hue ='sepal width (cm)')
```

```python
sns.pairplot(df,hue ='petal width (cm)')
```

### Dist Plot
A "dist plot" typically refers to a distribution plot, which is a plot that shows the distribution of a univariate (single variable) dataset.

It provides a visual representation of the distribution of values and can help identify patterns such as skewness, central tendency, and the presence of outliers.

```python
sns.distplot(df['sepal length (cm)'])
```

```python
sns.distplot(df['sepal length (cm)'],kde=False,bins=10)
```

# Categorical Plots
Seaborn also helps us in doing the analysis on Categorical Data points. In this section we will discuss about

1. boxplot
2. violinplot
3. countplot
4. bar plot

#### For categorical plot we are using tips dataset 

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

### Box Plot
A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset based on summary statistics. 

It provides a visual summary of the central tendency, spread, and skewness of the data, as well as the presence of outliers.

```python
sns.boxplot('smoker','total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",data=df)
```

# Violin Plot
A violin plot is a method of plotting numerical data and visualizing the distribution of the data across different levels of one or more categorical variables. 

It combines aspects of a box plot and a kernel density plot to provide a richer understanding of the data distribution.

```python
sns.violinplot(x="total_bill", y="day", data=df,palette='rainbow')
```
